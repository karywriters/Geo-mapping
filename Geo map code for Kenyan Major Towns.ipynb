```python
data = pd.read_csv('C:/Users/KARISA/Downloads/Skills gap.csv', index_col='County')
data = pd.DataFrame(data)
print(data)
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Path to the shapefile directories
Kenyan_boundaries_path = "C:/Users/KARISA/Downloads/Shapefile"
# Load the shapefiles
Kenyan_boundaries = gpd.read_file(Kenyan_boundaries_path)
Kenyan_boundaries


```

```python
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame with the 'Share' column indexed by county names
# Merge the shapefile data with your DataFrame based on county names
merged = Kenyan_boundaries.merge(data, how='left', left_on='county', right_index=True)
merged = merged.dropna()
merged
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame with the 'Share' column indexed by county names
# Merge the shapefile data with your DataFrame based on county names
merged = Kenyan_boundaries.merge(data, how='left', left_on='county', right_index=True)
merged = merged.dropna()
merged

```

```python
import matplotlib.pyplot as plt

# Plot the map of shaded counties
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
merged.plot(column='Average of skills_g', cmap='Oranges', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True, label=True)

# Plot the boundaries of Kenya without shading on top of the shaded map
Kenyan_boundaries.boundary.plot(ax=ax, color='black', linewidth=1)

# Add title and adjust plot settings
plt.title('Skills Gap by Major Cities in Kenya')
ax.axis('off')

# Create a legend for county names
merged.apply(lambda x: ax.annotate(text=x['county'], xy=x['geometry'].centroid.coords[0], horizontalalignment='center', fontsize=8), axis=1)

# Show the plot
plt.show()
```
